# Comparativa DBSCAN

## Preparación de ambiente

### Carga de módulos

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
import cufflinks as cf
import pandas as pd

cf.go_offline()
```

## Dataset

### Generación

```python
X, y = make_moons(n_samples=400, noise=0.075, random_state=0)
```

```python
moons = pd.DataFrame(X)
```

```python
moons
```

### EDA

```python
moons.set_index(0).iplot(kind="scatter", mode="markers", x=0, y=1)
```

## Modelado

### KMeans

```python
model = KMeans(n_clusters=2, n_jobs=-1)
```

```python
clusters = model.fit_predict(moons[[0, 1]])
```

```python
moons["cl"] = [str(x) for x in clusters]
```

```python
moons.iplot(kind="scatter", mode="markers", x=0, y=1, categories="cl")
```

### DBSCAN

```python
DBSCAN?
```

```python
model = DBSCAN(eps = 0.15)
```

```python
clusters = model.fit_predict(moons[[0, 1]])
```

```python
moons["cl"] = [str(x) for x in clusters]
```

```python
moons.iplot(kind="scatter", mode="markers", x=0, y=1, categories="cl")
```
